```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
from PIL import Image
from io import BytesIO

search_term = 'flower'#test
num_images_to_download = 200

url = rf'https://www.google.no/search?q={search_term}&client=opera&hs=cTQ&source=lnms&tbm=isch&sa=X&safe=active&ved=0ahUKEwig3LOx4PzKAhWGFywKHZyZAAgQ_AUIBygB&biw=1920&bih=982'

page = requests.get(url).text

soup = BeautifulSoup(page, 'html.parser')

listofimg = []

for raw_img in soup.find_all('img'):
    link = raw_img.get('src')

    if link and link.startswith("https://"):
        listofimg.append(link)

# Create a directory to save the images
output_directory = 'downloaded_images'
os.makedirs(output_directory, exist_ok=True)

# Download and save the images
for i, listofimg in enumerate(listofimg[:num_images_to_download]):
    response = requests.get(listofimg, stream=True)
    image_data = response.content

    # Use Pillow to determine the image format
    image = Image.open(BytesIO(image_data))
    image_format = image.format.lower() if image.format else 'jpeg'

    filename = os.path.join(output_directory, f'image_{i + 1}.{image_format}')

    with open(filename, 'wb') as file:
        file.write(image_data)

    if i + 1 == num_images_to_download:
        break

print(f"{num_images_to_download} images downloaded and saved to '{output_directory}' directory.")

#Problem - only 20 images were downloaded due to Google's restrictions on webscraping
```
